```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
#Make a new colunm for episode count
df.loc[0]
```

```python
def extract_episode (txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
           data = data + i
        if i == "(":
            check = True 
        
df['Episode'] = df['Title'].apply(extract_episode)
df['Episode'] = df['Episode'].str.replace(" eps","")
df['Episode'] = df['Episode'].astype(int)
df.head()
```

```python
#Make a new column for time stamp
def extract_timestamp(txt):
    check = False
    data = ""
    for i in range (len(txt)):
        if txt[i] == ")":
            for n in range (i+1, i+20):
                data += txt[n] 
            return data    
        
df['total time stamp'] = df['Title'].apply(extract_timestamp)
df.head()
```
