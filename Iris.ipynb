```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#checking any null values in dataset
iris=sns.load_dataset("iris")
iris.head()
#iris.to_excel(r"C:\Users\haris\Downloads\iris.xlsx", index=False)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.boxplot(x="species", y="sepal_length", data=iris, ax=axes[0,0])
axes[0,0].set_title("Sepal Length by Species")

sns.boxplot(x="species", y="sepal_width", data=iris, ax=axes[0,1])
axes[0,1].set_title("Sepal Width by Species")

sns.boxplot(x="species", y="petal_length", data=iris, ax=axes[1,0])
axes[1,0].set_title("Petal Length by Species")

sns.boxplot(x="species", y="petal_width", data=iris, ax=axes[1,1])
axes[1,1].set_title("Petal Width by Species")

plt.tight_layout()
plt.show()
```

```python
iris.hist(figsize=(10,8))
plt.show()
```

```python
sns.pairplot(data=iris,hue="species",diag_kind="kde")
plt.show()
```

```python
iris_setosa=iris[iris["species"]=="setosa"]
sns.histplot(data=iris_setosa,x="petal_length",bins=20,kde=True)
plt.title("Setosa")
plt.show()
```

```python
sns.displot(iris, x="sepal_length", bins=10, kde=True, col="species",height=3.7,aspect=0.8)
plt.show()
```

```python
g = sns.FacetGrid(iris, col="species")
g.map_dataframe(sns.kdeplot, x="sepal_width",color="r")
plt.show()
```

```python
iris_mean=iris.groupby("species").agg({"petal_length":["mean","max","min","std"]})
iris_mean.plot(kind="line",figsize=(8,6))
```

```python
corr=iris.corr(numeric_only=True)
sns.heatmap(data=corr,annot=True,cmap="Set2")
plt.show()

""" Commonly Used `cmap` Options

### 1. Sequential (good for ordered data, low  high)
- `"Blues"`  light to dark blue
- `"Greens"`  light to dark green
- `"Oranges"`  light to dark orange
- `"Purples"`  light to dark purple
- `"Reds"`  light to dark red
- `"YlGnBu"`  yellow  green  blue
- `"YlOrRd"`  yellow  orange  red
- `"BuPu"`  blue  purple
- `"GnBu"`  green  blue

---

### 2. Diverging (good for data centered around zero)
- `"coolwarm"`  blue  white  red
- `"bwr"`  blue  white  red (simpler)
- `"seismic"`  blue  white  red (strong contrast)
- `"RdBu"`  red  blue
- `"PiYG"`  pink  yellow  green
- `"PRGn"`  purple  green
- `"BrBG"`  brown  blue-green

---

### 3. Qualitative / Miscellaneous
- `"Set1"`, `"Set2"`, `"Set3"`  categorical palettes
- `"Pastel1"`, `"Pastel2"`  soft pastel colors
- `"tab10"`, `"tab20"`  distinct categorical colors
- `"Accent"`  bold accent colors"""
```
