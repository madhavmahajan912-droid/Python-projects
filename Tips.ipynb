```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# -------------------------------
# Relational Plots
# -------------------------------
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter Plot")
plt.show()
```

```python
sns.lineplot(data=tips, x="size", y="tip", hue="sex")
plt.title("Line Plot")
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", kind="scatter")
plt.title("Relational Plot (Scatter)")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Bar Plot")
plt.show()
```

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.title("Count Plot")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Violin Plot")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Strip Plot")
plt.show()
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Swarm Plot")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="bar", hue="sex")
plt.title("Cat Plot")
plt.show()
```

```python
sns.boxenplot(data=tips, x="day", y="total_bill")
plt.title("Boxen Plot")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram + KDE")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.title("KDE Plot")
plt.show()
```

```python
sns.ecdfplot(data=tips, x="total_bill")
plt.title("ECDF Plot")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kde=True)
plt.title("Displot")
plt.show()
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()
```

```python
sns.clustermap(corr, cmap="coolwarm", annot=True)
plt.title("Cluster Map")
plt.show()
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Regression Plot")
plt.show()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("LM Plot")
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.suptitle("Pair Plot", y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex", row="time", hue="day")
g.map(sns.scatterplot, "total_bill", "tip").add_legend()
plt.show()
```

```python
g = sns.PairGrid(tips, vars=["total_bill", "tip"], hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```
