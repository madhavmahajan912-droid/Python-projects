```python
df=nf.groupby("release_year")["title"].size().to_frame(name="count")
df_filtered=df[(df.index>=2016) & (df.index<=2021)]
df_filtered.T
```

```python
df_1=nf.groupby(["director","release_year"]).size().to_frame(name="count")
year=df_1.index.get_level_values("release_year")
df_filtered_1=df_1[(year>=2020) & (year<=2021)]
df_filtered_1.sort_values("count",ascending=False)
```

```python
a=sns.histplot(nf,x="release_year",bins=30,kde=True)
plt.show()
```

```python
top_director = nf['director'].value_counts().nlargest(10)
sns.barplot(x=top_director.values, y=top_director.index)
plt.title("Top 10 directors on Netflix by movies count")
plt.xlabel("Count")
plt.ylabel("Directors")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(nf,x="rating")
plt.title("Rating wise movies count")
plt.show()
```

```python
pie=nf.groupby("type").size().to_frame(name="count")
label=[f"{i}\n{j}" for i,j in zip(pie.index,pie.values)]
plt.pie(pie["count"],labels=label,autopct="%1.1f%%")
plt.title("Movies vs TV Shows")
plt.show()
```

```python
top_countries = nf['country'].value_counts().nlargest(10)
a=sns.lineplot(x=top_countries.values, y=top_countries.index, palette="coolwarm")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```
