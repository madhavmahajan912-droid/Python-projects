```python
plt.figure(figsize=(14,10))
corr=student.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

```python
g = sns.FacetGrid(student,row="sex",col="school")
g.map_dataframe(sns.histplot, x="G3", kde=True)
plt.show()
```

```python
g = sns.FacetGrid(student,col="school",height=4)
g.map_dataframe(sns.scatterplot, x="absences", y="G3")
g.add_legend()
plt.show()
```

```python
mjob_count = student["Mjob"].value_counts()
fjob_count = student["Fjob"].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

mjob_count.plot(kind="bar", ax=axes[0])
axes[0].set_title("Mother Job Count")

fjob_count.plot(kind="bar", ax=axes[1])
axes[1].set_title("Father Job Count")

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(6, 4))

sns.lineplot(data=student, x="studytime", y="G1", label="G1",errorbar=None)
sns.lineplot(data=student, x="studytime", y="G2", label="G2",errorbar=None)
sns.lineplot(data=student, x="studytime", y="G3", label="G3",errorbar=None)
plt.title("Effect of Study Time on Grades")
plt.xlabel("Study Time")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 3))
sns.boxplot(data=student, x="Mjob", y="G3")
plt.title("Student Performance by Mother's Job")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=student, x="sex", y="G3")
plt.title("G3 Distribution by Gender (Violin Plot)")
plt.show()
```

```python
grade_mean = student[["G1","G2","G3"]].mean()

plt.figure(figsize=(6,4))
grade_mean.plot(marker="o")
plt.title("Average Grade")
plt.xlabel("Exam")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(data=student, x="G2", y="G3")
plt.title("Regression: G2 vs G3")
plt.show()
```

```python
student["pass"] = student["G3"] >= 10

plt.figure(figsize=(5,4))
sns.countplot(data=student, x="pass")
plt.title("Pass vs Fail Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.ecdfplot(data=student, x="G3", hue="sex")
plt.title("ECDF of G3 by Gender")
plt.xlabel("G3 Score")
plt.ylabel("Proportion")
plt.show()
```
